# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
arr = np.zeros(10, dtype=int)
arr[4] = 1
arr = arr.reshape(2,5)
print(arr)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
arr = arr[::-1]
print(arr)
print(f'Четные элементы {arr[arr % 2 == 0]}')
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
#arr = np.random.randint(0,9,size=(3,3))
arr = np.arange(0,9).reshape(3,3)
print(arr)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
arr = np.random.rand(4,3,2)
print(arr)
print('--------')
print('minimum arr value =', np.min(arr))
print('max arr value =', np.max(arr))
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = np.random.rand(6,4)
arr2 = np.random.rand(4,3)
arr = arr1.dot(arr2)
print('matrix multiply =', arr)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.rand(7, 7)
print(arr)
print('----------')
min_val = np.min(arr)
max_val = np.max(arr)
print('min value arr=', min_val)
print('max value arr=', max_val)
print('average value arr=', np.mean(arr))
print('standard deviation arr=', np.std(arr))

# Проверяем, не равны ли min и max
if max_val == min_val:
    arr = np.zeros_like(arr)  # Заполняем массив нулями
else:
    arr = (arr - min_val) / (max_val - min_val)

print('normalized arr = \n', arr)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
print(tips.head(100))
```

### Узнайте сколько всего строчек и колонок в данных

```python
print(f'total rows {tips.shape[0]}, total columns {tips.shape[1]}')
```

### Проверьте есть ли пропуски в данных

```python
print('empty in data set\n', tips.isnull().sum() > 0)
print('count empty data in data set\n', tips.isnull().sum())
```

### Посмотрите на распределение числовых признаков

```python
tips.hist(bins=15, figsize=(10, 6))
```

### Найдите максимальное значение 'total_bill'

```python
print(tips['total_bill'].max())
```

### Найдите количество курящих людей

```python
print(tips[tips['smoker'] == 'Yes'].shape[0])
```

Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print('avarage total bill by day = ', tips.groupby('day')['total_bill'].mean())
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
grather_total_bill = tips[tips['total_bill'] > tips['total_bill'].median()]
print('total bill grather mean', grather_total_bill)
print('mean tip by sex = ', grather_total_bill.groupby('sex')['tip'].mean())
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'Yes':1, 'No':0})
print(tips.head(100))
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.pyplot.hist(tips['total_bill'], bins=20, edgecolor='black')
plt.pyplot.title('Гистограмма распределения признака "total_bill"')
plt.pyplot.xlabel('Total Bill')
plt.pyplot.ylabel('Частота')
```

Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

### Постройте pairplot

```python
sns.pairplot(tips)
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.histplot(data=tips, x='tip', hue='time', kde=True, bins=20, multiple='stack')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```
